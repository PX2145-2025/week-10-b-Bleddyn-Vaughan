```python
import matplotlib.pyplot as plt 
import numpy as np
from numpy.fft import fft, ifft
```

```python
#1

N=64

fs = 1000  # Hz
T = 1/fs   # sampling period
t = np.arange(N) * T
print(t)

f = 1 # frequency in Hz
x = np.cos(2*np.pi*f*t)  #create x array 
plt.plot(x,t)
plt.plot(x,t, 'o')  
plt.title('Single cosine wave sampled at {} points'.format(N))
plt.xlabel('time')
plt.ylabel('x(N)')
plt.grid()
```

```python
#2
N=64
tau = 10.0 # time length
t = np.linspace(0, tau, N, endpoint=False) # creates N points to cover tau seconds
print(t)

f = 1 # frequency in Hz
x = np.cos(2*np.pi*f*t)  #create x array 
plt.plot(x)
plt.plot(x, 'o') 
plt.title('Single cosine wave sampled at {} points'.format(N))
plt.xlabel('point number')
plt.ylabel('x(N)')
plt.grid()


```

```python
X = fft(x)/N # fft and normalise 
plt.plot(np.abs(X),'o')
plt.title('FFT of above cosine wave sampled at '+str(N)+' points') 
plt.xlabel('point number')
plt.ylabel(r'$\mid X(N)\mid$')
plt.grid()
print(X)
```

```python
#3
def Spectrum1(x, Fs):
    '''
    One-sided amplitude spectrum of x(t).
    '''
    N = len(x) # length of the signal
    frq = (float(Fs)/N)*np.arange(0, N/2+1) # one-sided frequency range 
    X = fft(x)/N # fft computing and normalization
    X = X[0:N//2+1] # take just positive values 
    
    
   
    return frq , X

N=64
tau = 10.0 # time length
t = np.linspace(0, tau, N, endpoint=False) # creates N points to cover tau seconds
print(t)
f = 0.2 # frequency in Hz
x = np.cos(2*np.pi*f*t)  #create x array 

frq, X = Spectrum1(x, f) #computing spectrum
plt.figure(figsize=(8,4))
plt.plot(frq, X, marker='o')
plt.title("One-Sided FFT of Cosine Wave")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

#the amplitude is only 0.5 since this only covers half the spectrum there is another peak in the negative spectrum if full spectrum was plotted would be amplitude 1 
```

```python
#6
N=64
tau = 10 # time length
t = np.linspace(0, tau, N, endpoint=False) # creates N points to cover tau seconds
print(t)
fs=10 # frequency in Hz
x = np.cos(2*np.pi*f*t)  #create x array 

f_list = [0.20, 0.22] #0.2 and 0.22 frequencies
phi = np.pi/4 #phase  45 degrees

for f0 in f_list:
    #signal with phase
    x = np.cos(2*np.pi*f0*t + phi)

    #compute FTT
    frq, X = Spectrum1(x, fs)
    #plot real and imaginary part
    plt.figure(figsize=(10,4))
    plt.plot(frq, np.real(X), 'b', label='Real(X)')
    plt.plot(frq, np.imag(X), 'r', label='Imag(X)')
    plt.title(f'Real and Imaginary parts of FFT, f0={f0} Hz, ={phi:.2f} rad')
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.grid(True)
    plt.show()
    #plot of magnitude and phase
    plt.figure(figsize=(10,4))
    plt.subplot(2,1,1)
    plt.plot(frq, np.abs(X), 'k')
    plt.ylabel("|X|")
    plt.title(f'Magnitude and Phase, f0={f0} Hz')
    plt.grid(True)

    plt.subplot(2,1,2)
    plt.plot(frq, np.angle(X), 'm')
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Phase (rad)")
    plt.grid(True)
    plt.show()
    
```
